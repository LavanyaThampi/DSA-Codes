**Line Plot**

```python
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 2, 22, 22, 10]
plt.plot(x, y)
plt.xlabel("days")
plt.ylabel("counts")
plt.show()
```

**Line plot axis control**

```python
import numpy as np
```

```python
x = np.arange(0,500)
y = np.cos(x)
plt.plot(x, y)
plt.xlim(0,100)
plt.show()
```

## **Multi Line plots**
- Labels and Legends

```python
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 3, 2, 22, 22, 10]
y2 = [1, 7, 5, 4, 3, 8]
plt.plot(x, y1, c="g", label="location1")
plt.plot(x, y2, c="r", label="location2")
plt.xlabel("days")
plt.ylabel("counts")
plt.legend(loc = "lower right")
plt.show()
```

**Bar Chart**

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
plt.bar(x, item_1_sales, label="item1", color="g")
plt.xlabel("days")
plt.ylabel("counts")
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
item_2_sales = [ 3, 4, 1, 6, 5]
plt.bar(x, item_1_sales, label="item1", color="b")
plt.bar(x, item_2_sales, alpha = 0.5, label="item2", color="g")
plt.xlabel("days")
plt.ylabel("counts")
plt.savefig("bar_chart.png", dpi=300)
plt.legend()
plt.show()
```

**Pie chart**

```python
countries = ["USA", "India", "UAE", "UK"]
population = [123, 20, 30, 40]
plt.pie(population, labels=countries, autopct="%1.2f%%")
plt.show()
```

**Histogram**

```python
import pandas as pd
data = pd.read_csv("auto-mpg.csv")
data.head()
```

```python
plt.hist(data["mpg"], color="green", rwidth=0.8, density=True)
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Density")
plt.show()
```

**Scatter Plot**

```python
plt.figure(figsize=(5, 3.5))
plt.scatter(data["weight"], data["mpg"], c="k", marker="*")
plt.title("MPG VS Weight")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

**Sub Figures**

```python
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 1)
plt.scatter(data["weight"], data["mpg"], c="k")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.subplot(1, 2, 2)
plt.scatter(data["displacement"], data["mpg"], c="b")
plt.xlabel("displacement")
plt.ylabel("MPG")
plt.show()
```

**Box Plot**

```python
a = np.random.normal(170, 1, 20000)
q1 = np.quantile(a, 0.25)
q3 = np.quantile(a, 0.75)
iqr = q3-q1
lower_bound = q1-(1.5*iqr)
upper_bound = q3+(1.5*iqr)
sns.boxplot(a)
```
