Saving Numpy Arrays
you can save arrays in both binary .npy format and compressed .npz format

```python
import numpy as np
```

```python
array = np.array([1,2,3,4,5,6])
np.save("array.npy", array)
print("Array saved as 'array.npy'")
```

```python
array1 = np.array([1,2,3])
array2 = np.array([4,5,6])
np.savez("arrays.npz", a1=array1,a2=array2)
print("Arrays saved as 'arrays.npz'")
```

```python
loaded_array = np.load("array.npy")
print("loaded_array:",loaded_array)
loaded_arrays = np.load("arrays.npz")
print("Array1:",loaded_arrays["a1"])
print("Array2:",loaded_arrays["a2"])
```

working with Audio Files

```python
import IPython.display as ipd
ipd.Audio('sample.wav')
```

```python
import librosa
```

```python
audio_data, sample_rate = librosa.load('sample.wav')
print(sample_rate)
```

```python
print(audio_data)
```

```python
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(6,4))
librosa.display.waveshow(audio_data, sr=sample_rate)
plt.title("Audio Waveform")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.show()
```
